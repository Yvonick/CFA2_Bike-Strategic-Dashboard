```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from datetime import datetime as dt
import seaborn as sns
```

```python
%%time
# Importing the Bike dataframe

#Importing the result into a new DF

# Using pyarrow to read the CSV to save on RAM
df_cleaned = pd.read_csv(
    "Output/bike_data_cleaned_2022.csv",
    engine="pyarrow",                 # faster & lower memory
    dtype_backend="pyarrow",          # compact Arrow dtypes for ints/strings
    parse_dates=["started_at"],       # avoids huge object strings
)

df_cleaned.columns
```

```python
# Set theme for the plots
# Palette: lane green, asphalt gray, Hudson blue, safety red
palette = 'crest'

sns.set_theme(
    context="notebook",
    style="darkgrid",
    palette=palette)
```

```python
# Bar Chart of the top 20 starting stations

# Create a dataframe with the number of rides for each Starting Station
df_start = pd.DataFrame(df_cleaned.groupby(['start_station_name'])['ride_id'].count()).reset_index()
df_start.rename(columns={'ride_id':'rides'}, inplace = True)

# Isolating the top20
df_start20 = df_start.nlargest(20, 'rides')

# Adjusting the color palette to display a continuous gradient adjusted to the size of the chart
# build colors from a continuous colormap using the rides values
cmap = sns.color_palette('crest', as_cmap=True)
norm = plt.Normalize(df_start20['rides'].min(), df_start20['rides'].max())
chartcolors = [cmap(norm(v)) for v in df_start20['rides'].to_numpy()]

# Plotting the data in a barchart
sns.barplot(
    data = df_start20, 
    x = 'rides', 
    y = 'start_station_name',
    hue = 'start_station_name',
    palette = chartcolors
)
plt.title("Top 20 Starting Stations")


# Displaying the chart

plt.show()
```

```python
# Dual plot of Daily Bike Trips VS Daily Average Temperature

# Import the dataframe from the previous exercise (+ clean up the date format)
df_ridetavg = pd.read_csv('Output/Ride Count VS Weather - NYC 2022.csv')

df_ridetavg['date'] = pd.to_datetime(df_ridetavg['date'], format='%Y-%m-%d', errors='coerce')
df_ridetavg = df_ridetavg.set_index('date').sort_index()

# Creating the twin plot

# Choosing a color for the 2 types of data
tripscolor = '#2AA876'
tavgcolor = '#B22222'

# Setting up the first subplot
fig,ax = plt.subplots(figsize=(10, 5))
sns.lineplot(data = df_ridetavg['rides'], color = tripscolor)
ax.set_xlabel("Year 2022") # set x-axis label
ax.set_ylabel("Daily Bike Trips", color = tripscolor) # set y-axis label

# Twinning it with the second subplot
ax2 = ax.twinx()
sns.lineplot(data = df_ridetavg['avgTemp'], ax=ax2, color = tavgcolor)
ax2.set_ylabel("Average temperatures", color = tavgcolor)

# Naming and displaying the twin plot
plt.title('Daily Bike Trips VS Daily Average Temperature - NYC 2022')
plt.show()
```

```python
%%time
# Let's make a box plot of the trip distance aginst the categorical variable "rideable_type" that differentiates between electric and classic bike

# First we need to calculate the trip distance
# We have the GPS coordinates of the start and end points
# We can make an estimation of the trip distance (as birds fly and assuming Earth is a circle)
def haversine_km(lat1, lon1, lat2, lon2, R=6371.0088):
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return 2*R*np.arctan2(np.sqrt(a), np.sqrt(1 - a))

df_cleaned['est_distance_km'] = haversine_km(df_cleaned['start_lat'], df_cleaned['start_lng'], df_cleaned['end_lat'], df_cleaned['end_lng'])
# df_cleaned.head()

```

```python
%%time
# Box plot the distance vs bike type

# The dataframe is too large to be fully plotted so we'll sample it
df_cleaned['est_distance_km'] = df_cleaned['est_distance_km'].astype('float32')

N = 1_000_000 # A sample of 1 million is more than enough to be representative
n = len(df_cleaned)
rng = np.random.default_rng(None)
idx = rng.choice(n, size=min(N, n), replace=False)

# Create the sampled dataframe with only the needed columns
df_sampled = df_cleaned.loc[idx, ['rideable_type','est_distance_km']].dropna().copy()

# Sanity check on the sample representativeness
print(df_sampled['rideable_type'].value_counts())

# Plot the box chart
plt.figure(figsize=(9,5))
sns.boxplot(
    x = 'rideable_type', 
    y = 'est_distance_km' , 
    data = df_sampled,
    showfliers=False,
)
plt.show()
```

This boxchart shows us that:
- classic bikes are used for shorter trips and electric bikes for longer trips
- the trip distance is mostly comprised between 0 and 6 km, regardless of the type of bike, with most trips being less than 3km

This is not very surprising but an important user behaviour assumption to confirm. This could help decide where to put more electric bike slots (e.g. remote stations that commute with the city center) and where we can afford to have fewer (e.g. cluster of stations that do mostly short commutes).

The obvious limitation of this analysis is that we estimated the trip distance using the GPS coordinates, which does not take into account the actual topology of the city. It's still probably a good indicator, as statistically this should correlate strongly with the actual trip distance. But it's probably quite shorter than the actual trip distance.

This assumption could be tested by sampling N trips, running the start/destination through an itinerary calculator and comparing the distance with our long/lat based estimation. If we find a strong correlation, we can even calculate the ratio between actual distance and estimated distance and use it in our analysis.

# FacetGrid
FacetGrid would be useful in this latter analyis, by showing the comparative volume of classic bike VS electric bike, in addition to their distance distribution.

```python
grid = sns.FacetGrid(df_sampled, col = 'rideable_type')

# Zoom in on the relevant range
grid.set(xlim=(0, 10))
grid.map(sns.histplot, 'est_distance_km', bins = 15)
```

We can compare the absolute volumes of classic VS electric bike :
- For estimated distances under 2.5km, the volume of classic bikes is significantly larger than electric bikes
- For longer distances, we see more electrical bikes in absolute numbers

We also get a better view of the distribution of trip distances for both types, and we can see they have roughly the same exponentially decreasing shape. The main difference being the higher sharpness of the decrease for the classic bike.
